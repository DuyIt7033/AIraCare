```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import matplotlib.pyplot as plt

model = tf.keras.models.load_model('E:/lv/DenseNet1/best_model.h5')

# image_folder = 'D:/x-ray_split/test/0_normal'
# image_folder = 'D:/x-ray_split/test/1_covid19'
image_folder = 'D:/x-ray_split/test/2_Pneumonia'

class_labels = ['Normal', 'Covid19', 'Pneumonia']
class_counts = {'Normal': 0, 'Covid19': 0, 'Pneumonia': 0}

IMAGE_SIZE = (224, 224)

for filename in os.listdir(image_folder):
    if filename.endswith('.jpg') or filename.endswith('.png') or filename.endswith('.jpeg'):
        img_path = os.path.join(image_folder, filename)

        # Load ảnh và chuyển đổi thành numpy array
        img = load_img(img_path, target_size=IMAGE_SIZE)
        img_array = img_to_array(img) / 255.0  
        img_array = np.expand_dims(img_array, axis=0)  

        # Dự đoán 
        predictions = model.predict(img_array)
        predicted_class = np.argmax(predictions, axis=1)[0]

        # Lấy nhãn của lớp dự đoán
        class_label = class_labels[predicted_class]
        class_counts[class_label] += 1

print("Số lượng ảnh thuộc mỗi lớp:")
for label, count in class_counts.items():
    print(f"{label}: {count}")
```
